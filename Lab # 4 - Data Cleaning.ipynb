### Candy Data 

```python
# One line to open and read an excel file into a data frame data structure, make the first row the header,
# then close the file. 
df = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx', encoding = "ISO-8859-1")

# Print the first couple of lines to ensure that everything is correct
df.head()
```

```python
# Dropping the last column because by looking at the metadata (the original survey) this last row is not an
# actual survey question
df.drop(' [York Peppermint Patties] Ignore', axis = 1, inplace = True)

# Printing the info to see if there is one less column now that I have dropped one. We started with 123, 
# so now we should (and do) have 122
df.info ()
```

```python
# The next thing I wanted to do was to rename the headers, because they are excessively wordy, and don't always make sense. 
# Because I didn't have time to do every one (there are over 100), I did so with just the first five. If I had more time,
# I would do all of them

df.rename(columns = {
    'Timestamp' : 'Time' ,
    'Are you going actually going trick or treating yourself?' : 'ToT?' ,
    'Your gender:' : 'Gender' ,
    'How old are you?' :'Age' ,
    'Which country do you live in?' : 'Country' , 
    },  inplace = True )
```

```python
# I noticed inconsistensies and incorrect spellings of countries. For example usa vs US vs USA. All of these be 
# recognized as different countries even though they are all the same. I made the entire country column lowercase so that 
# would eliminate one difference. 
df["Country"] = df["Country"].str.lower()
df.head ()


```

```python
#Then I took a couple of variations of USA to all become the same. If I had more time, 
# I would apply this to every country that had this issue, and address every variation. 
df.replace(to_replace = "us", value = "usa", inplace = True)

df.head ()
# Now, the first four "usa" entries are all the same. By changing the case and replacing us to usa. 
```

```python
# I found that the columns with open ended questions had data that was not sortable because the surveyed people 
# could enter whatever they wanted. This made me think that they were not worth keeping, so I deleted age. This is because
# for age, they did not have to select their age, but could type out whatever they wanted. There were a lot of answers that 
# were not numbers. 

df.drop('Age', axis = 1, inplace = True)
df.head()


```

```python
# Lastly for this dataset, I decided that if there was a blank in one of the columns that allowed someone to rate a candy, 
# I would replace it with "undecided". This would show that there were unanswered pieces of that question, but would not 
# disturb the data for the people who answered. It would make it clear how many people didn't answer it for whatever reason
# I only did this for one column, the [100 Grand Bar] column, but if I had time I'd do it for all of them

df.replace(to_replace = "", value = "undecided", inplace = True)

df.head()
```

### Attendance Data

```python
# Changed every header name so that it was clear/understandable for me, removed the multiple headerrows, and displayed to ensure
# it was working correctly. Opened and read an excel file into a data frame data structure, then closed it. 

headern = ["states", "combined_ada", "combined_ada2", "combined_hours", "combined_hours2", "combined_days_in_year", 
              "combined_days_in_year2", "combined_hours_in_year","combined_hours_in_year2", "elementary_ada", 
              "elementary_ada2", "elementary_hours_in_day", "elementary_hours_in_day2",
               "secondary_ada", "secondary_ada2", "secondary_hours_in_a_day", "secondary_hours_in_a_day2"]

df2 = pd.read_excel("attendance.xls", skiprows=3, names=headern)

df2.head()
```

```python
# Dropped rows 66 - 69 (this is the index on the excel file, which is why the code doesnot match up with these numbers)
# because they are index rows and not actual data values. Checked info before and after to ensure that 
# the number of rows decreased 
df2.drop(df2.index[[61,62,63,64]], inplace = True )
```

```python
# I then dropped the empty rows that were in betweeen states and empty for no reason 
df2.drop(df2.index[[6, 12, 18, 24, 30, 36, 42, 48]], inplace = True )
```

```python
# I then wanted to fill all empty or valueless points. From the index I found that ‡ means insufficient data and † means
# not applicable. So I replaced these with those words so someone analyzing the data can easily understand without
# having to look at an index
df2.replace(to_replace = "†", value = "not applicable", inplace = True)
df2.replace(to_replace = "‡", value = "insufficient data", inplace = True)
df2.replace(to_replace = "(†)", value = "not applicable", inplace = True)
df2.replace(to_replace = "(‡)", value = "insufficient data", inplace = True)

# I printed Deleware's row to make sure that it replaced the right symbols with the right strings I wanted. 
print(df2.loc[[9]])
```
